```python
file = '2023/geoTiffs/prueba2.tif'

with rs.open(file) as src:
    # Lee los datos de las tres primeras bandas (RGB)
    rgb_image = src.read([1, 2, 3])

    # Normaliza los valores de los pxeles para que estn en el rango [0, 1]
    rgb_image = rgb_image / rgb_image.max()

    # Visualiza la imagen RGB
    pyplot.imshow(rgb_image.transpose(1, 2, 0))
    pyplot.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Define the path to the PNG file
png_file_path = "2023/geoTiffs/prueba2.png"  

# Check if the file exists
if os.path.exists(png_file_path):
    # Load the PNG image using matplotlib
    img = mpimg.imread(png_file_path)
    img = img / img.max()

    # Display the image using matplotlib
    plt.imshow(img)
    plt.axis('off')  # Turn off axis
    plt.show()
else:
    print(f"File {png_file_path} not found.")
```
